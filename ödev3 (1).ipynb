```python
matris = np.random.randint(0, 10, (3, 2))
print("Matris:",matris)

satir = len(matris)
sutun = len(matris[0])

transpoz = []

for j in range(sutun):
    yeni_satir = []
    for i in range(satir):
        yeni_satir.append(matris[i][j])
    transpoz.append(yeni_satir)

print("Transpoz:",np.array(transpoz))
```
